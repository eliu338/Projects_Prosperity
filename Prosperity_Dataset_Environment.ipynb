```python
import numpy as np

def compute_autoregressive_coefficients_fixed(time_series):
    if len(time_series) < 5:
        raise ValueError("Time series must have at least 5 points")

    coefficients = []

    for i in range(4, len(time_series)):
        X = np.array([
            time_series[i-4],
            time_series[i-3],
            time_series[i-2],
            time_series[i-1]
        ]).reshape(-1, 4)

        Y = np.array([time_series[i]])

        # Adding a column of ones for the intercept in X
        X = np.hstack([np.ones((X.shape[0], 1)), X])

        # Use np.linalg.lstsq to compute the coefficients in a more robust way
        coeffs, _, _, _ = np.linalg.lstsq(X, Y, rcond=None)

        # Store the coefficients (excluding the intercept)
        coefficients.append(coeffs[1:])

    # Calculate the average of each coefficient across all computations
    average_coefficients = np.mean(coefficients, axis=0)

    print("Average Coefficients:", average_coefficients)
    return coefficients, average_coefficients

# Attempt the example usage again
coefficients, average_coefficients = compute_autoregressive_coefficients_fixed(list(starfruit_mid_prices))
```

```python
import numpy as np

def compute_autoregressive_coefficients_with_regularization(time_series, lambda_reg=0.01):
    if len(time_series) < 5:
        raise ValueError("Time series must have at least 5 points")

    coefficients = []

    for i in range(4, len(time_series)):
        X = np.array([
            [1],  # Intercept
            [time_series[i-4]],
            [time_series[i-3]],
            [time_series[i-2]],
            [time_series[i-1]]
        ]).T

        Y = np.array(time_series[i])

        # Add regularization term to X^T * X
        XtX = np.dot(X.T, X)
        XtX_reg = XtX + lambda_reg * np.eye(X.shape[1])

        # Compute the inverse of the regularized XtX
        XtX_reg_inv = np.linalg.inv(XtX_reg)

        # Compute coefficients: (X^T * X + lambda*I)^-1 * X^T * Y
        coeffs = np.dot(XtX_reg_inv, np.dot(X.T, Y))

        coefficients.append(coeffs)

    average_coefficients = np.mean(coefficients, axis=0)

    print("Average Coefficients (Including Intercept):", average_coefficients)
    return coefficients, average_coefficients



coefficients, average_coefficients = compute_autoregressive_coefficients_with_regularization(list(starfruitdiffs[1:]))
```

```python
import re

# A function to parse the log file and extract MM MIDPOINT, NEW MID PRICE, and timestamps
def extract_data_from_log(file_path):
    mm_midpoint_list = []
    new_mid_price_list = []
    timestamps = []

    # Regular expressions to find MM MIDPOINT, NEW MID PRICE, and timestamp
    mm_midpoint_pattern = re.compile(r"MM MIDPOINT (\d+\.\d+)")
    new_mid_price_pattern = re.compile(r"NEW MID PRICE (\d+\.\d+)")
    timestamp_pattern = re.compile(r'"timestamp": (\d+)')

    with open(file_path, 'r') as file:
        log_entry = ''
        for line in file:
            # Accumulate lines until we complete a JSON entry
            log_entry += line
            if line.startswith('}'):
                # Extract MM MIDPOINT
                mm_midpoint_match = mm_midpoint_pattern.search(log_entry)
                if mm_midpoint_match:
                    mm_midpoint_list.append(float(mm_midpoint_match.group(1)))

                # Extract NEW MID PRICE
                new_mid_price_match = new_mid_price_pattern.search(log_entry)
                if new_mid_price_match:
                    new_mid_price_list.append(float(new_mid_price_match.group(1)))

                # Extract timestamp
                timestamp_match = timestamp_pattern.search(log_entry)
                if timestamp_match:
                    timestamps.append(int(timestamp_match.group(1)))

                # Reset log entry for the next JSON block
                log_entry = ''

    return timestamps, mm_midpoint_list, new_mid_price_list

# Extract data from the log file
timestamps, mm_midpoint_list, new_mid_price_list = extract_data_from_log("/content/drive/MyDrive/Finalfruitslog.log")


```

```python
import pandas as pd

def process_data(file_path):
    data = pd.read_csv(file_path, delimiter=',')
    products = ['STRAWBERRIES', 'CHOCOLATE', 'ROSES', 'GIFT_BASKET']
    filtered_data = data[data['product'].isin(products)]
    grouped = filtered_data.groupby(['timestamp', 'product'])

    def calculate_midpoint(group):
        best_bid = group['bid_price_1'].max()
        best_ask = group['ask_price_1'].min()
        midpoint = (best_bid + best_ask) / 2
        return round(midpoint)

    results = grouped.apply(calculate_midpoint).reset_index()
    results.columns = ['Timestamp', 'Product', 'Processed Value']

    return results

def calculate_spread(results):
    pivot = results.pivot(index='Timestamp', columns='Product', values='Processed Value')
    pivot['Spread'] = pivot['GIFT_BASKET'] - 6 * pivot['STRAWBERRIES'] - 4 * pivot['CHOCOLATE'] - pivot['ROSES'] - 375
    return pivot

def trade_positions(spreads, data, x=15):
    max_positions = {'STRAWBERRIES': 350, 'CHOCOLATE': 250, 'ROSES': 60, 'GIFT_BASKET': 60}
    transactions = []
    pnl = 0

    for timestamp, row in spreads.iterrows():
        action = 'Hold'
        amount = 0

        if row['Spread'] > x:  # Buy Gift Baskets
            best_ask = data.loc[(data['timestamp'] == timestamp) & (data['product'] == 'GIFT_BASKET'), 'ask_price_1'].min()
            if pd.notna(best_ask):
                amount = max_positions['GIFT_BASKET']
                pnl -= amount * best_ask
                action = f'Buy {amount} Gift Baskets'
        elif row['Spread'] < -x:  # Sell Gift Baskets
            # print(data.timestamp)
            best_bid = data.loc[(data['timestamp'] == timestamp) & (data['product'] == 'GIFT_BASKET'), 'bid_price_1'].max()
            if pd.notna(best_bid):
                amount = max_positions['GIFT_BASKET']
                pnl += amount * best_bid
                action = f'Sell {amount} Gift Baskets'

        transactions.append({
            'Timestamp': timestamp,
            'Action': action,
            'P&L': pnl
        })

    return pd.DataFrame(transactions)

file_path = '/content/drive/MyDrive/r4log.csv'
processed_results = process_data(file_path)
spread_results = calculate_spread(processed_results)
transaction_results = trade_positions(spread_results, processed_results)
print(transaction_results)
```

```python
import pandas as pd

def process_data(file_path):
    data = pd.read_csv(file_path, delimiter=',')
    products = ['STRAWBERRIES', 'CHOCOLATE', 'ROSES', 'GIFT_BASKET']
    filtered_data = data[data['product'].isin(products)]

    grouped = filtered_data.groupby(['timestamp', 'product'])

    def calculate_details(group):
        # Find the maximum bid price and corresponding volume across all bid columns
        best_bid_price = group[['bid_price_1', 'bid_price_2', 'bid_price_3']].max(axis=1).max()
        bid_cols = ['bid_price_1', 'bid_price_2', 'bid_price_3']
        bid_vol_cols = ['bid_volume_1', 'bid_volume_2', 'bid_volume_3']
        for i in range(3):  # Check all three bid price columns
            if best_bid_price in group[bid_cols[i]].values:
                best_bid_volume = group[group[bid_cols[i]] == best_bid_price][bid_vol_cols[i]].iloc[0]
                break

        # Find the minimum ask price and corresponding volume across all ask columns
        best_ask_price = group[['ask_price_1', 'ask_price_2', 'ask_price_3']].min(axis=1).min()
        ask_cols = ['ask_price_1', 'ask_price_2', 'ask_price_3']
        ask_vol_cols = ['ask_volume_1', 'ask_volume_2', 'ask_volume_3']
        for j in range(3):  # Check all three ask price columns
            if best_ask_price in group[ask_cols[j]].values:
                best_ask_volume = group[group[ask_cols[j]] == best_ask_price][ask_vol_cols[j]].iloc[0]
                break

        midpoint = (best_bid_price + best_ask_price) / 2
        return pd.Series({
            'Best Bid Price': best_bid_price,
            'Best Bid Volume': best_bid_volume,
            'Best Ask Price': best_ask_price,
            'Best Ask Volume': best_ask_volume,
            'Midpoint': round(midpoint)
        })

    results = grouped.apply(calculate_details).reset_index()
    return results

def calculate_spread_and_merge(results):
    # Pivot to get midpoints as columns for each product per timestamp
    pivot = results.pivot(index='timestamp', columns='product', values='Midpoint')

    # Calculate the spread for each timestamp
    pivot['Spread'] = (pivot['GIFT_BASKET'] - 6 * pivot.get('STRAWBERRIES', 0)
                       - pivot.get('ROSES', 0) - 4 * pivot.get('CHOCOLATE', 0) - 375)

    # Reset the index to flatten the DataFrame
    pivot.reset_index(inplace=True)

    # Merge the spreads into the processed results
    final_results = pd.merge(results, pivot[['timestamp', 'Spread']], on='timestamp', how='left')

    # Save the final results to a CSV file
    final_results.to_csv("final_results_with_spreads.csv", index=False)
    return final_results

# Example usage
file_path = "/content/r4log.csv"  # Change this path to your actual file location
processed_results = process_data(file_path)
final_results = calculate_spread_and_merge(processed_results)
print(final_results)
```

```python
import pandas as pd

def run_trading_simulation(file_path, x=15, max_position=60, min_position=-60):
    # Load the data
    data = pd.read_csv(file_path)

    # Filter data for "GIFT_BASKET" only
    data = data[data['product'] == 'GIFT_BASKET']

    # Initial positions and PnL
    position = 0
    pnl = 0

    # Print initial setup
    print(f"Initial Setup -> Max Position: {max_position}, Threshold x: {x}")

    # Iterate through each row in the DataFrame
    for index, row in data.iterrows():
        spread = row['Spread']
        best_bid_price = row['Best Bid Price']
        best_bid_volume = row['Best Bid Volume']
        best_ask_price = row['Best Ask Price']
        best_ask_volume = row['Best Ask Volume']
        mid_price = row['Midpoint']

        # Debug output for each row processed
        print(f"Processing -> Timestamp: {row['timestamp']}, Spread: {spread}, Position: {position}, PnL: {pnl}")

        if spread > x and position < max_position:
            # Calculate how much we can buy
            trade_amount = min(max_position - position, best_ask_volume)
            if trade_amount > 0:
                position += trade_amount
                pnl -= trade_amount * best_ask_price
                print(f"Action -> Bought {trade_amount} units at ask price {best_ask_price}, New Position: {position}, Updated PnL: {pnl}")
        elif spread < -x and position > -max_position:
            # Calculate how much we can sell
            trade_amount = min(abs(-position + min_position), best_bid_volume)
            if trade_amount > 0:
                position -= trade_amount
                pnl += trade_amount * best_bid_price
                print(f"Action -> Sold {trade_amount} units at bid price {best_bid_price}, New Position: {position}, Updated PnL: {pnl}")

    # print(f"Final Position: {position}, Final PnL: {pnl}")
    if position < 0:
        pnl -= position * mid_price
    elif position > 0:
        pnl += position * mid_price
    return pnl

# Set the path to your CSV file
file_path = "/content/final_results_with_spreads.csv"
final_pnl = run_trading_simulation(file_path)
print(f"Final PnL: {final_pnl}")
```
